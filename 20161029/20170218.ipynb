```python
#'mom, anna, racecar' 'race car'

def check_palingdrome(word):
    #reverse word
    #print 'raw word:', word
    reversed_word = word[::-1]
    #print 'reversed word:', reversed_word
    if word == reversed_word:
        #print word, 'is a palingdrome.'
        return True
    else:
        #print word, 'is not a palingdrome.'
        return False

w1 = 'raceca'
check_palingdrome(w1)
```

```python
sentence = 'anna has a beautiful racecar'
words = sentence.split()
for word in words:
    result = check_palingdrome(word)
    print word, result
```
